## Document Parsing Stress Test: Identified Challenges 
For this exploration, two specific pages from the PUE (Rules for Electrical Installation) were deliberately selected. These pages serve as an excellent stress test because they contain a high density of complex typographical elements, layout inconsistencies, and common parsing artifacts.

Upon initial extraction using `docling`, we identified several structural and formatting issues that require further attention (solutions will be addressed in subsequent steps):



###  Run-in (Inline) Headings
The source document uses inconsistent typography for numbered list items. While some items (e.g., "4.") are formatted as isolated structural blocks, others (e.g., "5.", "6.") are "run-in" headings—bold text that immediately precedes the paragraph body without a line break. The visual parser struggles with this inconsistency. It correctly tags the isolated blocks as Headings but incorrectly classifies the run-in headings as standard Paragraphs, which breaks the logical hierarchy of the extracted Markdown.

### Borderless "Ghost" Tables
The document contains small, implicit tables (e.g., matching Nominal Voltage to Resistance) that rely purely on visual whitespace alignment rather than explicit grid lines. Because the parser's table-detection algorithms are optimized for structural grids, it fails to recognize these elements as tabular data. Instead, it flattens the columns into continuous, unstructured text strings.

### Undecoded Mathematical Formulas
The extraction pipeline correctly detects the presence of complex, two-dimensional mathematical equations (e.g., the formula for computing insulation resistance). However, out-of-the-box, it fails to decode the mathematical syntax (like fractions and subscripts). Instead of returning garbled text, it safely falls back to a placeholder tag: <!-- formula-not-decoded>.

### Hyphenation and Line Break Artifacts
Standard typographic line breaks and word wrap in the original PDF result in trailing hyphens and unwanted spaces in the extracted text (e.g., "сра -батывания"). These artifacts break word boundaries, which degrades the semantic quality of the text and can negatively impact downstream NLP tasks or vector search accuracy.

### Textual "Spaghetti" from Embedded Schematics (OCR Overreach)
When processing electrical single-line diagrams and grounding plans, the parser successfully detects the presence of an image but subsequently applies overly aggressive OCR to the internal labels (e.g., "L1", "L2", "PE"). Because these labels derive their meaning entirely from their spatial context (their position next to specific wires or contacts), extracting them in a linear, top-to-bottom fashion generates meaningless, disconnected text strings that pollute the output.

### Captions Swallowed by Images
Instead of extracting the figure caption (e.g., "Рис. 1.7.4...") and its legend as distinct textual metadata linked to the image, the layout parser overestimates the spatial boundaries of the schematic. It draws the bounding box too generously, effectively swallowing the caption text into the image asset itself. As a result, this critical descriptive text is locked inside the visual block and lost from the logical markdown text flow.

### Multi-page Table Fragmentation
When a complex table spans across a pagination boundary (e.g., Table 1.8.1 split between pages 4 and 5), the parser fails to maintain continuous structural integrity. It interprets the physical page break as a definitive table termination, resulting in two fragmented tables. Furthermore, repeating headers and continuation markers (like "Окончание табл. 1.8.1") are treated as standalone text or erroneously merged into data cells, severely complicating downstream data extraction.

## Objective Assessment of Parser Capabilities

Based on the stress test, the identified parsing artifacts fall into three distinct categories regarding how effectively `docling` can resolve them:

### 1. Resolvable via Configuration (Native Capabilities)
These issues can be fixed by tuning the `PipelineOptions` within the parser configuration:
* **OCR Overreach (Spaghetti Text):** Can be eliminated by explicitly disabling OCR for elements classified as `Picture` or `Figure`.
* **Undecoded Formulas:** Can be resolved by enabling the formula extraction flags and ensuring the appropriate OCR dependencies are active.

### 2. High Regression Risk (Trade-offs)
These issues can theoretically be addressed by the parser, but adjusting the parameters is highly risky:
* **Borderless "Ghost" Tables & Swallowed Captions:** Tuning the sensitivity of the TableFormer or adjusting Bounding Box thresholds might fix specific edge cases but will likely introduce severe regressions (e.g., misclassifying standard text as tables or cropping legitimate image assets).

### 3. Unresolvable by Visual Parsers (Semantic Constraints)
These issues are physically impossible for spatial layout parsers to solve reliably because they require semantic and linguistic understanding:
* **Run-in Headings:** The parser cannot logically split a single continuous paragraph based solely on font-weight.
* **Hyphenation Artifacts:** The parser lacks the morphological dictionary required to distinguish between line-break hyphens and legitimate compound words.
* **Multi-page Table Fragmentation:** The parser evaluates pages in isolation and cannot maintain context across pagination boundaries.


### Possible Architectural Decision: The Hybrid "Agentic" Pipeline
Attempting to force the visual parser to solve semantic problems is an anti-pattern. To build a robust, production-ready ingestion pipeline (`doc_agent`), we will adopt a two-stage architecture:

1. **Stage 1 (Physical Parsing):** Configure `docling` strictly for spatial layout, table extraction, and clean image cropping (disabling internal OCR for schematics).
2. **Stage 2 (Semantic Normalization):** Introduce an **LLM Post-Processor Agent** with a zero-temperature, highly constrained system prompt to fix run-in headings, heal fragmented tables, and resolve hyphenation artifacts using its natural language understanding.

### Exploration Outcome: The Limits of Physical Parsing
We attempted to resolve the visual artifacts by aggressively tuning the `PdfPipelineOptions` (enabling `do_formula_enrichment`, forcing `generate_picture_images`, and completely disabling `do_ocr`). 

**Results:**
* **Partially Successful (Image Isolation):** Disabling OCR (`do_ocr = False`) successfully prevented the parser from hallucinating "spaghetti text" out of schematic labels. The images were correctly isolated and exported to disk.
* **Failed (Formulas):** Enabling formula enrichment did not resolve the formula placeholders, exposing the limitations of the local layout parser without heavy, specialized mathematical OCR models.
* **Unresolved (Semantic Artifacts):** Multi-page tables and swallowed captions remained untouched. 

**Conclusion:** We have hit the ceiling of what pure heuristic/spatial parsing can reliably achieve. To build a robust extraction pipeline, we must freeze the physical parsing (Stage 1) in its current state and delegate the semantic normalization (LaTeX decoding, table merging, and caption restoration) to an LLM Post-Processor Agent (Stage 2).
